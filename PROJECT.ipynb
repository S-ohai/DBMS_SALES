EXPLORATORY DATA ANALYSIS

```python
#pictorial representation based on order id and quantity
profitability_by_Order_ID = df.groupby('Order ID')['Quantity'].sum()
plt.figure(figsize=(10,10))
profitability_by_Order_ID.plot(kind='barh', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by Order_ID')
plt.ylabel('Order ID')
plt.xlabel('Quantity')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```

```python
#pictorial representation based on order id and amount
profitability_by_Order_ID = df.groupby('Order ID')['Amount'].sum()
plt.figure(figsize=(10,10))
profitability_by_Order_ID.plot(kind='barh', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by Order_ID')
plt.ylabel('Order ID')
plt.xlabel('Amount')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```

```python
#pictorial representation based on payment mode and profit
profitability_by_payment_mode = df.groupby('PaymentMode')['Profit'].sum()
plt.figure(figsize=(10, 6))
profitability_by_payment_mode.plot(kind='line', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by Payment Mode')
plt.xlabel('Payment Mode')
plt.ylabel('Total Profit')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```

```python
#pictorial representation based on category
profitability_by_Category = df.groupby('Category')['Profit'].sum()
plt.figure(figsize=(20, 6))
profitability_by_Category.plot(kind='barh', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by Category')
plt.ylabel('Category')
plt.xlabel('Total Profit')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```

```python
#pictorial representation based on sub-category
profitability_by_Sub_Category = df.groupby('Sub-Category')['Profit'].sum()
plt.figure(figsize=(40, 6))
profitability_by_Sub_Category.plot(kind='area', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Profit')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```

```python
# Assuming 'Profit' is already in the correct numerical format.
# Sum or mean can be calculated depending on the desired metric.
profitability_by_State = df.groupby('State')['Profit'].sum()
plt.figure(figsize=(20, 6))
profitability_by_State.plot(kind='bar', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by State')
plt.xlabel('State')
plt.ylabel('Total Profit')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```

```python
#pictorial representation based on city
profitability_by_City = df.groupby('City')['Profit'].sum()
plt.figure(figsize=(20, 30))
profitability_by_City.plot(kind='pie', color='skyblue')  # Use a visually pleasing color
plt.title('Profitability by City')
plt.xlabel('City')
plt.ylabel('Total Profit')  # Or 'Average Profit', depending on the calculation
plt.xticks(rotation=45)  # Rotate x-axis labels if they are too lengthy or overlapping
plt.show()
```
